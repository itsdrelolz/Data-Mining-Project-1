### Apple Quality Dataset

### This dataset is a collection attributes related to apples. These attributes include Size, Weight, Sweetness, Crunchiness, Juiciness, Ripeness, Acidity, and Quality. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Clean data 
#### Drop null values from Size, Weight, Sweetness, Crunchiness, Juiciness, Ripeness, and Quality. 
#### Remove A_id column

```python
df = pd.read_csv('apple_quality.csv')

df.info()

df = df.dropna()
df.drop('A_id', axis=1, inplace=True)
```

# Checking amount of good vs. bad apples

### There is around a 50/50 split between good and bad apples in this dataset

```python
df['Quality'].value_counts()

# quality df 
quality_counts = df['Quality'].value_counts()


bars = plt.bar(quality_counts.index, quality_counts)

plt.xlabel('Quality')
plt.ylabel('Count')
plt.title('Bar Chart of Quality')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom')

plt.show()
```

```python
df.plot(x='Size', y='Weight', kind='scatter')
plt.title('Size vs Weight Scatter Plot')
```

```python
df.plot(x='Sweetness', y='Crunchiness', kind='scatter')
plt.title('Sweetness vs Crunchiness Scatter Plot')
```

```python
df.plot(x='Ripeness', y='Juiciness', kind='scatter')
plt.title('Ripeness vs Juiciness Scatter Plot')
```

# Isolating the apples that are only considered "good" in order to see what the average values look like

```python
# create new df that contains only "good" quality apples
good_quality_df = df[df['Quality'] == 'good']
#calculate mean for all values in the "good" dataframe
average_values_good_quality = good_quality_df[['Sweetness', 'Crunchiness', 'Acidity', 'Sweetness', 'Crunchiness', 'Acidity', 'Size', 'Weight', 'Juiciness', 'Ripeness']].mean()


plt.bar(average_values_good_quality.index, average_values_good_quality)




plt.xlabel('Attributes')
plt.ylabel('Average Value')
plt.title('Average Sweetness, Crunchiness, Acidity, Sweetness, Crunchiness, Acidity, Size, Weight, Juiciness, and Ripeness of "Good" Quality Apples')

plt.ylim(-1, 1)

plt.show()
```

# Isolating the apples that are only considered "bad" in order to see what the average values look like

```python
# create new df that contains only "bad" quality apples
good_quality_df = df[df['Quality'] == 'bad']
# calculate the mean for all values in the "bad" df
average_values_good_quality = good_quality_df[['Sweetness', 'Crunchiness', 'Acidity', 'Size', 'Weight', 'Juiciness', 'Ripeness']].mean()


plt.bar(average_values_good_quality.index, average_values_good_quality)




plt.xlabel('Attributes')
plt.ylabel('Average Value')
plt.title('Average Sweetness, Crunchiness, Acidity,Sweetness, Crunchiness, Acidity, Size, Weight, Juiciness, and Ripeness of "Bad" Quality Apples')

plt.ylim(-1, 1)

plt.show()
```
